```python
import pandas as pd
from time import time
from sqlalchemy import create_engine
```

```python
pd.__version__
```

```python
df = pd.read_csv("yellow_tripdata_2021-01.csv", nrows=100)
df
# df.head(100)
```

```python
df_schema = pd.io.sql.get_schema(df, "yellow_tripdata_schema")
print(df_schema)
```

```python
# create sqlalchemy engine for the ny_taxi postgres container
conn_nytaxi = "postgresql://root:root12345@localhost:5432/ny_taxi"
engine = create_engine(conn_nytaxi)
```

```python
# pickup datetimes are currently in TEXT, these needs to be converted accordingly
df["tpep_pickup_datetime"] = pd.to_datetime(df["tpep_pickup_datetime"])
df["tpep_dropoff_datetime"] = pd.to_datetime(df["tpep_dropoff_datetime"])

df_schema = pd.io.sql.get_schema(df, "yellow_tripdata_schema", con=engine)
print(df_schema)
```

```python
# implement dataframe iterator s=to split large datasets to chunks
df_iter = pd.read_csv("yellow_tripdata_2021-01.csv", iterator=True, chunksize=100000)
```

```python
# # insert data by chunks
# while True:
#     t_start = time()
    
#     df_curr_chunk = next(df_iter)

#     df_curr_chunk["tpep_pickup_datetime"] = pd.to_datetime(df_curr_chunk["tpep_pickup_datetime"])
#     df_curr_chunk["tpep_dropoff_datetime"] = pd.to_datetime(df_curr_chunk["tpep_dropoff_datetime"])

#     # finally append data
#     df_curr_chunk.to_sql(name="yellow_taxi_data", con=engine, if_exists="append")

#     t_end = time()
#     print(f"chunk inserted...duration: {t_end - t_start:.4f} seconds")
```

```python
# flag for iter 1
is_first_iter = True

for df_chunk in df_iter:

    t_start = time()
    
    # pickup datetimes are currently in TEXT, these needs to be converted accordingly
    df_chunk["tpep_pickup_datetime"] = pd.to_datetime(df_chunk["tpep_pickup_datetime"])
    df_chunk["tpep_dropoff_datetime"] = pd.to_datetime(df_chunk["tpep_dropoff_datetime"])

    # create table schema in postgres (table without data)
    if is_first_iter:
        df_chunk.head(0).to_sql(name="yellow_taxi_data", con=engine, if_exists="replace")
    
    df_chunk.to_sql(name="yellow_taxi_data", con=engine, if_exists="append")

    t_end = time()
    print(f"chunk inserted...duration: {t_end - t_start:.4f} seconds")

    is_first_iter = False

print(f"Pipeline completed...")
```
